# 2. Data cleaning

```python
df[df['language']!='old']
```

```python
df=df[df['language']!='old'].reset_index()
df.head()
```

```python
df['duration'] [:10]
```

```python
# duration to seconds
durations=[]
for i in df['duration']:
  lst=i.split(":")
  if(len(lst)==3):
   durations.append(int(lst[0])*60*60 + int(lst[1])*60+int(lst[2]))
  elif(len(lst)==2):
    durations.append(int(lst[0])*60 + int(lst[1]))
  else:
    durations.append(int(lst[0]))
df['duration']=durations

```
